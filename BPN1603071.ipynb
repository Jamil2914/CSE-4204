# **Assigning the Class Label to each corresponding input entry**

```python
no_of_output_node = 5
total_class = pow(2,no_of_output_node)  
listLabels = [] #list containing corresponding class label to each entry
listFeatureEntries = [] #list containing entry values
for iterator in range(no_of_inputs):
    listFeatureEntries.append(iterator)
val = 0
a=0
for iterator in range(no_of_inputs):
    if a>0 and (a%2==0):
        val+=1
    if val==total_class:
            val = 0
    listLabels.append(val)
    a+=1
```

# **Dataset preparation by splitting Training and testing sets**

```python
train_data = []
train_label = []
test_data = []
test_label = []
train_data_limit = 70*no_of_inputs/100
for i in range(0, no_of_inputs):
    if i<int(train_data_limit):
        train_data.append(listFeatureEntries[i])
        train_label.append(listLabels[i])

    else:
        test_data.append(listFeatureEntries[i])
        test_label.append(listLabels[i])


```

# **Function declaration for Right Shift,BitValue and Conversion of feature value**

```python
def RSH(value,n):
    return value>>n
```

```python
def bitValue(Val,N):
    return (RSH(Val,N))&1
```

```python
def decimalToBinary(val):
    ret = np.zeros(total_input_bit)
    for i in range(total_input_bit):
        ret[i] = bitValue(val,i)  # the bit value on the ith bit

    return ret
```

# **DECLARATION OF VARIABLES and WEIGHT INITIALIZATION**

```python
train_data = np.array(train_data)
train_label = np.array(train_label)
test_data = np.array(test_data)
test_label = np.array(test_label)
```

```python
NodesLayer1 = total_input_bit 
NodesLayer2 = 14
NodesLayer3 = no_of_output_node
```

```python
Wij = np.zeros((NodesLayer1, NodesLayer2))
Wjk = np.zeros((NodesLayer2, NodesLayer3))
```

```python
DelWij = np.zeros((NodesLayer1, NodesLayer2))
DelWjk = np.zeros((NodesLayer2, NodesLayer3))
```

```python
Uhj = np.zeros(NodesLayer2)
Uok = np.zeros(NodesLayer3)
```

```python
DelUhj = np.zeros(NodesLayer2)
DelUok = np.zeros(NodesLayer3)
```

```python
for i in range(NodesLayer1):
    for j in range(NodesLayer2):
        Wij[i][j] = random.random()

for i in range(NodesLayer2):
    Uhj[i] = 0.5
    for j in range(NodesLayer3):
        Wjk[i][j] = random.random()

for i in range(NodesLayer3):
    Uok[i] = 0.5

Oaj = np.zeros(NodesLayer2)
Oak = np.zeros(NodesLayer3)

eta1 = 0.9
eta2 = 0.9
th = 0.5
```

```python
print("Weight Matrix first layer to second layer:", Wij)

print("Weight Matrix Second layer to third layer:", Wjk)
```

```python
# import module
from tabulate import tabulate
 
# assign data
mydata = [
    [Wij.shape,
Wjk.shape,
DelWij.shape,
DelWjk.shape,
Uhj.shape,
Uok.shape,
DelUhj.shape,
DelWjk.shape,
Uhj.shape,
Uok.shape,
DelUhj.shape,
DelUok.shape]
]

# create header
head = ["Wij.shape","Wjk.shape","DelWij.shape","DelWjk.shape","Uhj.shape","Uok.shape","DelUhj.shape","DelWjk.shape","Uhj.shape","Uok.shape","DelUhj.shape","DelUok.shape"  ]
 
# display table
print(tabulate(mydata, headers=head, tablefmt="grid",numalign='center'))
```

```python

MATCHES = 0; tot_error = 0
err = np.zeros(5)

def BPN(inputfeatures, labels):
    global MATCHES, tot_error, err

    # Calculation of Net,Activ,Oaj(Input Layer to Hidden Layer)
    net = np.zeros(NodesLayer2)
    activ = np.zeros(NodesLayer2)
    K = 0.5
    A = 1

    for j in range(NodesLayer2):
        for i in range(NodesLayer1):
            net[j] += Wij[i][j]*inputfeatures[i]

        activ[j] = net[j] + Uhj[j]
        Oaj[j] = 1.0/(A + math.exp(-K*activ[j]))

    # Calculation of activOk and Netk from hidden layer to output layer
    net = np.zeros(NodesLayer3)
    activ = np.zeros(NodesLayer3)
    err = np.zeros(NodesLayer3)

    for k in range(NodesLayer3):
        for j in range(NodesLayer2):
            net[k] += Wjk[j][k]*Oaj[j]

        activ[k] = net[k] + Uok[k]
        Oak[k] = 1.0/(A + math.exp(-K*activ[k]))

    # Error Calculation (Track of matches by flag)
    FLAG = 1
    targ_feat = decimalToBinary(labels)
    for k in range(NodesLayer3):
        err[k] = targ_feat[k] - Oak[k]
        tot_error += (targ_feat[k] - Oak[k])**2

        
        if Oak[k] >= th:
            bitTocompare = 1
        if Oak[k] < th:
            bitTocompare = 0
        if bitTocompare != targ_feat[k]:
            FLAG = 0

    # Block 4 (Back Prop Starts)
    for j in range(NodesLayer2):
        for k in range(NodesLayer3):
            DelWjk[j][k] = eta2*K*err[k]*Oaj[j]*Oak[k]*(1-Oak[k])
            Wjk[j][k] += DelWjk[j][k]

    for k in range(NodesLayer3):
        DelUok[k] = eta2*K*err[k]*Oak[k]*(1-Oak[k])
        Uok[k] += DelUok[k]

    # Block 5
    for i in range(NodesLayer1):
        for j in range(NodesLayer2):
            sum = 0
            for k in range(NodesLayer3):
                sum += err[k]*Wjk[j][k]

            DelWij[i][j] = eta1*K*inputfeatures[i]*Oaj[j]*(1-Oaj[j])*sum
            Wij[i][j] += DelWij[i][j]

            DelUhj[j] = eta1*K*Oaj[j]*(1-Oaj[j])*sum
            Uhj[j] += DelUhj[j]
    return FLAG
```

```python

while (1):
    MATCHES = 0; tot_error = 0
    for i in range(train_data.shape[0]):
        feat = decimalToBinary(train_data[i])
        MATCHES += BPN(feat, train_label[i])
    head = ["Accuracy","Total Error","Oaj","Oak","ERR","Uok","Uhj"]
    Accuracy = MATCHES/train_data.shape[0]*100.0
    mydata=[[Accuracy,tot_error*0.5,Oaj,Oak,err,Uok,Uhj]]
    print(tabulate(mydata, headers=head, tablefmt="grid",numalign='center'))
    if tot_error*0.5<=1:
        break
```

```python
epoch = 10
for ep in range(epoch):
    MATCHES = 0; tot_error = 0

    for i in range(train_data.shape[0]):
        feat = decimalToBinary(train_data[i])
        MATCHES += BPN(feat, train_label[i])
    head = ["Epoch","Accuracy","Total Error","Oaj","Oak","ERR","Uok","Uhj"]
    Accuracy = MATCHES/train_data.shape[0]*100.0
    mydata=[[ep,Accuracy,tot_error*0.5,Oaj,Oak,err,Uok,Uhj]]
    print(tabulate(mydata, headers=head, tablefmt="grid",numalign='center'))


 
    
```
